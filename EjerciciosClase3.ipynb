ejercicio ifs

```python
año_de_lanzamiento = 1991
respuesta=int(input("¿Cuando se lanzo Python por primera vez?: "))

if año_de_lanzamiento == respuesta:
    print("Correcto")
else:
    if año_de_lanzamiento > respuesta:
        print("Te fuiste muy arriba")
    else:
        print("Te fuiste muy abajo")

print("Adios!")
```

```python
año_de_lanzamiento = 1991
correcto = False

while not (correcto):
  respuesta = int(input("En que año se lanzo Python?: "))

  if año_de_lanzamiento == respuesta:
      print("Correcto")
      correcto = True
  else:
      print("Intentalo nuevamente")

print("Adios!")
```

```python
def imprimir_argumentos(*args):
  for i in args:
    if type(i) == int:
      continue
    print(i)

  return

imprimir_argumentos("a", 1,"b", 2, "c", 3, "d",  4)
```
