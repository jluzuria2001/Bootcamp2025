```python
numero_dias=int(input("Ingrese el numero de dias: "))

num_años=numero_dias//365
num_semanas=(numero_dias%365)//7
num_dias_restante=(numero_dias%365)%7

print("años: "+ str(num_años))
print("semanas restantes: "+ str(num_semanas))
print("dias restantes: "+ str(num_dias_restante))
```
